```python
df['total_kids'] = df[['Kidhome','Teenhome']].sum(axis=1)
```

```python
df
```

```python
df['marital_Divorced'] = df['marital_Divorced'].replace({1:5, 0:0})
df['marital_Married'] = df['marital_Married'].replace({1:4, 0:0})
df['marital_Single'] = df['marital_Single'].replace({1:3, 0:0})
df['marital_Together'] = df['marital_Together'].replace({1:2, 0:0})
df['marital_Widow'] = df['marital_Widow'].replace({1:1, 0:0})
```

```python
df['marital_status'] = df[['marital_Divorced','marital_Married','marital_Single','marital_Together','marital_Widow']].sum(axis=1)
```

```python
df['marital_status_str'] = df['marital_status'].map({5:'Divorced',4:'married',3:'single',2:'together',1:'widow'})
```

```python
df.head()
```

```python
df['education_2n Cycle'] = df['education_2n Cycle'].replace({1:1, 0:0})
df['education_Basic'] = df['education_Basic'].replace({1:2, 0:0})
df['education_Graduation'] = df['education_Graduation'].replace({1:3, 0:0})
df['education_Master'] = df['education_Master'].replace({1:4, 0:0})
df['education_PhD'] = df['education_PhD'].replace({1:5, 0:0})
```

```python
df['education_status'] = df[['education_2n Cycle','education_Basic','education_Graduation','education_Master','education_PhD']].sum(axis=1) 
```

```python
df['accepted_campaigns'] = df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']].sum(axis=1)
```

```python
df[df['accepted_campaigns'] != 0].head()
```

```python
df['accepted_campaigns'] = (df['accepted_campaigns']!= 0).astype(int)
```

```python
df.corr()['accepted_campaigns'].sort_values(ascending=False)
```

```python
all_correlations = df.corr()
all_correlations = all_correlations[(all_correlations > 0.3) & (all_correlations < 1)]
sns.heatmap(all_correlations)
```

```python
all_correlations['accepted_campaigns'].sort_values(ascending=True)
```

By looking at the above features we can see that there is definitely correlation between how much people are spending and accepting campaigns.
If they are spending more money at the store they are more likely to accept a campaign.

```python
age_groups = [(20,30),(31,40),(41,50),(51,60),(61,70),(71,80)]

def assign_age(Age):
    for age_range in age_groups:
        if age_range[0] <=Age <=age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return 'unknown'

df['Age_group'] = df['Age'].apply(assign_age)
```

```python
df.head()
```

```python
 df[['Age_group','Age']].sort_values('Age_group', ascending=True)
```

```python
df_age_group = pd.DataFrame(df.groupby('Age_group')['accepted_campaigns'].count())
df_age_group = df_age_group.reset_index()
```

```python
df_age_group
```

```python
df_age_group['percentage'] = df_age_group.accepted_campaigns/df.shape[0]
df_age_group
```

```python
sns.barplot(x='Age_group',y='percentage',data=df_age_group)
plt.title('percentage of campaigns most selected by age group')
plt.show()
```

```python
#age segmentation : core audience accepting campaigns right now is 31 - 70 year olds.
```

```python
df_accepted_cam = df[df['accepted_campaigns'] != 0]
df_amt_total = pd.DataFrame(df_accepted_cam.groupby('Age_group')['MntTotal'].sum().reset_index())
df_amt_total
```
