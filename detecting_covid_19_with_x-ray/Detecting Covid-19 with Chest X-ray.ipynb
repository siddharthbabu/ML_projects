## Data preparation and cleaning

```python
import os
import cv2 
import itertools
import numpy as np 
import pandas as pd 
```

```python
covid_dir = "C:/Users/Suribabu/data_science/projects/detecting_covid_19_with_x-ray/COVID/"
non_covid_dir = "C:/Users/Suribabu/data_science/projects/detecting_covid_19_with_x-ray/non-COVID/"
data = []
```

```python
for dirtrain in os.listdir(covid_dir): 
    print(dirtrain)
    img = cv2.imread(covid_dir + dirtrain)
    img = cv2.resize(img, (32, 32))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = img.reshape(32, 32, 1)
    new_label = dirtrain.split('(')
    print(new_label[0])
    data.append([img, new_label[0]]) 
```

```python
for dirtrain in os.listdir(non_covid_dir): 
    print(dirtrain)
    img = cv2.imread(non_covid_dir + dirtrain)
    img = cv2.resize(img, (32, 32))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = img.reshape(32, 32, 1)
    new_label = dirtrain.split('(')
    print(new_label[0])
    data.append([img, new_label[0]]) 
```

In each iteration we:</br>

-read image </br>
-resize the image to matrix of 32x32</br>
-grayscaling image</br>
-reshape the image to 3D with one color channel (Gray)</br>
-append each image with label to the data list</br>

```python
import random
random.seed(20)
```

```python
random.shuffle(data)
```

```python
x, y = [], []
for e in data:
    x.append(e[0])
    y.append(e[1])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
x = np.array(x)
y = np.array(y)
y = y.reshape(-1,1)
enc = OneHotEncoder(handle_unknown='ignore').fit(y)
print(enc.categories_)
y = enc.transform(y).toarray()
print(f'Data   :   {str(x.shape)}')
print(f'Labels :   {str(y.shape)}')
```

```python
from sklearn.model_selection import train_test_split
#We then split the data into training and validation sets. We use 80% of the data for training and 20% for validation.
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.2)
```

## Building the model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(64, (4, 4), padding='same', activation=tf.nn.relu, 
    input_shape=(32, 32, 1)),
    tf.keras.layers.MaxPooling2D((2, 2), strides=(2,2)), Dropout(0.25),
    
    tf.keras.layers.Conv2D(128, (3,3), padding='same', activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D((2, 2), strides=(2,2)),
    Dropout(0.25),

    tf.keras.layers.Conv2D(128, (3,3), padding='same', activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D((2, 2), strides=(2,2)),
    Dropout(0.3),

    tf.keras.layers.Conv2D(128, (2,2), padding='same', activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D((2, 2), strides=(2,2)),
    Dropout(0.3),

    tf.keras.layers.Conv2D(256, (2,2), padding='same', activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D((2, 2), strides=(2,2)),
    Dropout(0.3),
                    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(2,  activation=tf.nn.softmax)
])
```

```python
#model summary
model.summary()
```

## training the model

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
hist = model.fit(x_train, y_train, epochs=100, validation_split=0.2, batch_size=64,verbose=1,shuffle=True)
```

## plotting training and vallidation accuracy

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

fig = plt.figure(figsize=(12, 6), dpi=80)
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

We can see that the loss is decreasing with each epoch. We can also see that the validation loss is decreasing with each epoch. This means that the model is not overfitting the training data.

```python
fig = plt.figure(figsize=(12, 6), dpi=80)
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

We can see that the accuracy is increasing with each epoch

## Evaluating the model

```python
loss_and_metrics = model.evaluate(x_test, y_test, verbose=2)
y_pred = model.predict(x_test).argmax(axis=1)
print(f'Test Loss     : {loss_and_metrics[0]}')
print(f'Test Accuracy : {loss_and_metrics[1]}')
     
```

```python
from sklearn import metrics

df = pd.DataFrame(
  data = metrics.confusion_matrix(np.argmax(y_test, axis=1), y_pred),
  columns = ['0', '1'],
  index = ['0', '1']
)
df
```
