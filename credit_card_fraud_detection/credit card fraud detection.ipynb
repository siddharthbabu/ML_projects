```python
#splitting the data
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score,classification_report
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#models used
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

### Model 1

```python
#Random forest classifier
rfc = RandomForestClassifier()
model = rfc.fit(X_train, y_train)
rfc_pred = model.predict(X_test)
```

```python
print("For Random Forest classifier") 
  
acc = accuracy_score(y_test, rfc_pred) 
print("The accuracy is {}".format(acc)) 
  
prec = precision_score(y_test, rfc_pred) 
print("The precision is {}".format(prec)) 
  
rec = recall_score(y_test, rfc_pred) 
print("The recall is {}".format(rec)) 
  
f1 = f1_score(y_test, rfc_pred) 
print("The F1-Score is {}".format(f1)) 
print("\nClassification Report:",classification_report(y_test, rfc_pred))
```

### Model 2

```python
#Logistic regression
lr = LogisticRegression()
model_2 = lr.fit(X_train, y_train)
lr_pred = model_2.predict(X_test)
```

```python
print("For Random Forest classifier") 
  
acc = accuracy_score(y_test, lr_pred) 
print("The accuracy is {}".format(acc)) 
  
prec = precision_score(y_test, lr_pred) 
print("The precision is {}".format(prec)) 
  
rec = recall_score(y_test, lr_pred) 
print("The recall is {}".format(rec)) 
  
f1 = f1_score(y_test, lr_pred) 
print("The F1-Score is {}".format(f1)) 

print("\nClassification Report:",classification_report(y_test, lr_pred))
```

### Model 3

```python
#Decision tree
dc = DecisionTreeClassifier()
model_3 = dc.fit(X_train, y_train)
dc_pred = model_3.predict(X_test)
```

```python
print("For Random Forest classifier") 
  
acc = accuracy_score(y_test, dc_pred) 
print("The accuracy is {}".format(acc)) 
  
prec = precision_score(y_test, dc_pred) 
print("The precision is {}".format(prec)) 
  
rec = recall_score(y_test, dc_pred) 
print("The recall is {}".format(rec)) 
  
f1 = f1_score(y_test, dc_pred) 
print("The F1-Score is {}".format(f1)) 
print("\nClassification Report:",classification_report(y_test, dc_pred))
```

### Conclusion

Looking at the classification reports we can conclude that:
Precision: Report 1 has the highest precision for class 1, indicating that it has the highest proportion of correct positive predictions among all positive predictions. This is important if minimizing false positives is a priority.

Recall: Report 3 has the highest recall for class 1, indicating that it has the highest proportion of correctly predicted positive instances out of all actual positive instances. This is important if capturing all positive instances is a priority.

F1-score: The F1-score is the harmonic mean of precision and recall, so it provides a balance between the two metrics. Report 1 has the highest F1-score for class 1, indicating a balance between precision and recall.

Accuracy: All reports have the same overall accuracy, but accuracy alone may not be the best metric for imbalanced datasets like fraud detection, where the number of non-fraudulent transactions far exceeds the number of fraudulent transactions.

In conclusion, Report 1 may be preferred due to its high precision. If the goal is to capture as many fraudulent transactions as possible, Report 3 may be preferred due to its high recall.
