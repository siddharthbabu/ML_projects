```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read all data from csv
df = pd.read_csv('imdb_movies.csv')
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#check for duplicates
df[df.duplicated()]
```

```python
#drop duplicates
df.drop_duplicates(inplace=True)
```

```python
#after dropping duplicates
df[df.duplicated()]
```

```python
df.dropna(subset=['genres'],inplace=True)
```

```python
df['profit'] = df['revenue'] - df['budget']
```

```python
df_genre = df[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','release_year','profit']]
```

```python
df_genre
```

```python
from pandas import Series, DataFrame
split = df_genre['genres'].str.split('|').apply(Series,1).stack()
```

```python
split.index = split.index.droplevel(-1)
split
```

```python
split.name = 'genres_split'
del df_genre['genres']
df_genre = df_genre.join(split)
```

```python
df_genre[df_genre['original_title']=='Jurassic World']
```

### Research questions
#### 1. Which genres are the most common?

```python
genres_count = pd.DataFrame(df_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending=False)
```

```python
genres_count
```

```python
genres_count['original_title'].plot.pie(title='movies per genre in %',autopct='%1.1f%%',figsize=(9,9))
```

```python
genres_count['original_title'].plot.barh(title='movies per genre count',color='DarkBlue',figsize=(10,9))
```

#### 2. Which genres have high avg.budget and revenue?

```python
genres_mean = df_genre.groupby('genres_split').mean()
pd.options.display.float_format = '{:2f}'.format
genres_mean
```

```python
genres_mean.sort_values('budget',ascending=False,inplace=True)
```

```python
genres_mean[['budget','revenue']].plot.barh(title='budget and revenue by genres',color=('DarkBlue','c'),figsize=(10,9))
```

#### 3. Which genres have high avg profit?

```python
genres_mean.sort_values(by='profit',ascending=True,inplace=True)
```

```python
genres_mean['profit'].plot.barh(title='profit by genre',color='DarkBlue',figsize=(10,9))
```

#### 4. which genre has the highest popularity?

```python
genres_mean.sort_values(by='popularity',ascending=True,inplace=True)
```

```python
genres_mean['popularity'].plot.barh(title='popularity by genre',color='DarkBlue',figsize=(10,9))
```

#### 4. which genres has the highest number of movies with an avg vote > 6?

```python
vote_fifty = df_genre[(df_genre['vote_count']>=50) & (df_genre['vote_average']>=8)]
```

```python
genre_vote = pd.DataFrame(vote_fifty.groupby('genres_split').vote_average.nunique()).sort_values('vote_average',ascending=True)
```

```python
genre_vote
```

```python
genre_vote['vote_average'].plot.barh(title='highest vote avg based on genre',color='DarkBlue',figsize=(10,9))
```

```python
df_counted = df[df['vote_count']>=50]
df_counted.corr(method='spearman')
```

```python
sns.regplot(x='vote_average',y='profit', data=df_counted, line_kws={'color':'red'})
```

From the above plot we can see that there isnt a lot of correlation between vote average and profit.

```python
#question 2
sns.regplot(x='popularity',y='profit', data=df_counted, line_kws={'color':'red'})
```

Looking at this plot we can conclude that poopularity increases profit

```python
#question 3
sns.regplot(x='budget',y='profit', data=df_counted, line_kws={'color':'red'})
```

From the above plot we can conclude that budget and profit have a decent correlation </br>
Hence we can conclude that higher budget can produce a higher profit

```python
#question 4
sns.regplot(x='budget',y='popularity', data=df_counted, line_kws={'color':'red'})
```

From the above plot we can conclude that budget and popularity have a decent correlation </br>
Hence we can conclude that higher budget will provide higher popularity

```python
df_genre = df_genre[['popularity','budget','revenue','original_title','runtime','release_year','vote_count','vote_average','profit','genres_split']]
```

```python
df_genre
```

```python
df_time_genre = pd.DataFrame(df_genre.groupby(['release_year','genres_split'])['profit'].mean())
df_time_genre
```

```python
final_genre = pd.pivot_table(df_time_genre, values='profit',index=['genres_split'],columns='release_year')
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.heatmap(final_genre, cmap='coolwarm',linewidth=.5)
plt.title('genres by profit per year')
```

From the above heatmap we can see certain genres like fantasy, adventure and science fiction had increased interest </br>
Animation had a good run in the early 60's and history and western had high profits in the 90's.
